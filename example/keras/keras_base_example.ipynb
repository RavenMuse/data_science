### kears 交互式操作

#### 方式1：纯keras backend

```python
from keras import backend as K
```

```python
# 写法1
session = K.get_session()
kcons = K.constant(value=[[1, 2], [3, 4]])
dot_res=K.dot(kcons,kcons)
dot_res.eval(session=session)
```

```python
# 写法2
kcons = K.constant(value=[[1, 2], [3, 4]])
dot_res=K.dot(kcons,kcons)
K.eval(dot_res)
```

#### 方式2 ：keras backend与tf混用

```python
import tensorflow as tf
from keras import backend as K
```

```python
# 写法1
sess = tf.Session()
Q_seq=tf.constant([[2.], [1.], [5.]])
sess.run(K.shape(Q_seq))
```

```python
# 写法2
Q_seq=tf.constant([[2.], [1.], [5.]])
res=tf.matmul(tf.transpose(Q_seq),Q_seq) 
K.eval(res)
```

```python
# 变量定义
init_normal=initializers.TruncatedNormal(stddev=0.02)
K.variable(init_normal(shape=[1,2],dtype=K.floatx()),name='w')
```
